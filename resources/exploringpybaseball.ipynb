```python
import os
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import pybaseball as pyball 
from pybaseball import statcast
from pybaseball import playerid_lookup
from pybaseball import batting_stats 
import datetime
```

```python
data = pd.DataFrame(batting_stats(2014,2021, qual=50))
#show columns info
data 
```

```python
pd.set_option('display.max_columns', None)

data.columns.tolist()
```

```python
#create view with only relevant columns for analysis
data_relevant_columns = data[['Age','PA','H','HR','R','RBI','SB','BB','SO','AVG','OBP','SLG','OPS','wOBA','WAR']]

#make a correlation heatmap for these columns, ignore any columns that don't have numeric data
plt.figure(figsize=(12, 10))
sns.heatmap(data_relevant_columns.corr(), annot=True, cmap='coolwarm')
plt.show()

```

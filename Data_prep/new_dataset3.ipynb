```python
import pandas as pd
```

```python
# load data from csv
data = pd.read_csv('../resources/clean_extended_data.csv')
```

```python
data.head()
```

```python
# keep Important columns IDfg SO, CSW%, LD, Events, wRAA, Barrels, RBI, IFFB, R, competitive_runs and drop others
neew_data = data[['IDfg', 'SO', 'CSW%', 'LD', 'Events', 'wRAA', 'Barrels', 'RBI', 'IFFB', 'R', 'competitive_runs']]



```

```python
# Load old data2.csv
data2 = pd.read_csv('../resources/data2.csv')
data2.head()
```

```python
# Keep important column from data2.csv 'next_year_hits' and drop others
new_data2 = data2[['IDfg','next_year_hits']]
```

```python
# Merge the two dataframes and match the IDfg column
merged_data = pd.merge(neew_data, new_data2, on='IDfg')
merged_data.head()
```

```python
# Save the merged data to a new csv file
merged_data.to_csv('../resources/data3.csv', index=False)
```
